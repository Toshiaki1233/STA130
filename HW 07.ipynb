1. There is only one independent variable (X) and one dependent variable (Y) in simple linear regression. Finding the best-fitting straight line that depicts the relationship between X and Y is the aim; this method is usually applied when there is only one influencing factor or the main element impacting the result. However, in order to forecast the dependent variable Y, Multiplie Linear Regression uses several independent variables (X1, X2,..., Xn). MLR captures a greater variety of influences on the dependent variable by incorporating numerous variables, which frequently results in more accurate predictions. 
2. A continuous variable is a quantitative variable that can take an infinite number of values within a range. Examples include age, temperature, or height. On the other hand, An indicator variable is a categorical variable that has been converted into binary form (0 or 1) to represent different groups or categories. For example, a variable that indicates gender might assign "0" for male and "1" for female. 
3. A Multiple Linear Regression (MLR) model can take into consideration both the difference between two groups and the impact of the continuous variable on the outcome by using an indicator variable in addition to the continuous variable. Two parallel lines, one for each group represented by the indicator variable and each with a different intercept but the same slope, are obtained from this MLR model. Assuming a constant rate of change with regard to the continuous variable, this configuration allows the model to capture a change in the baseline level of the outcome between the groups. Simple Linear Regression (SLR), on the other hand, models a single linear trend across all data without making any distinctions between groups, using simply a continuous variable.
4. In a Multiple Linear Regression model, the effect of a continuous variable on the outcome might vary between groups by including an interaction between the continuous and an indicator variable. In the absence of the interaction, the model just modifies the intercept for each group, applying the same slope for the continuous variable. However, each group exhibits a unique slope and intercept for the interaction, indicating that the continuous variable's influence varies by group. A more detailed model that represents different impacts across categories is made possible by this, which is useful for capturing interactions where trends change between groups.
5. Without making any assumptions about a continuous or linear trend, a Multiple Linear Regression model that is exclusively based on indicator variables from a non-binary categorical variable can capture variations in the results between categories. If "Region" is the categorical variable, for instance, and the categories are "North," "South," and "East," the model will designate a distinct intercept for each category, signifying the average result for that category in comparison to a baseline (such as "West"). Binary encodings, or dummy variables, are used in this method for all but one category, which acts as the reference. By considering categories as independent and unordered, the model is able to quantify various group effects. Each coefficient quantifies the difference in result for each category relative to the baseline.

> _The way an **interaction** actually appears and works in the context of the linear form of a **multiple linear regression** model may not necessarily be immediately intuitive, as it is actually in fact somewhat subtle and tricky. Of course, an **interaction** is when the relationship of one **predictor variable** with the **outcome variable** depends on the value of another different **predictor variable**, so the impact of one **predictor variable** changes based on the presence or magnitude of another **predictor variable**. But are you sure you know what this means in the context of the **linear form** of a **multiple linear regression** model?_
>
> - _Imagine you're selling smoothies, the taste of your smoothie depends on the ingredients included in the smoothie, and there are two ingredients (bananas and strawberries) in the smoothie._
>     - _Adding more bananas into the smoothie will of course increase the "banana" flavor of the smoothie, and vice-versa for strawberries..._
>     - _But is this "banana" influence on the flavor always constant, or can it change depending on the absolute amount of strawberries in the smoothie?_ 
>     - _If the "banana" flavor influence is constant and does not depend on the  absolute amount of strawberries in the smoothie, then there is no **interaction** and the **linear form** of the model is $\beta_b b_i + \beta_s s_i$ and the model is said to be only **additive**._
>     - _But if the "banana" flavor influence does depend on the absolute amount of strawberries in the smoothie, then there IS an **interaction** and the **linear form** of the model is $\beta_b b_i + \beta_s s_i + \beta_{bs} (b_i\times s_i)$ and the model is said to be **synergistic**._
>         
> _These **linear forms** show that either bananas and strawberries do not have any **synergistic interaction** and contribute to the flavor independently; or, they do have a **synergistic interaction** and there is an interesting interplay between bananas and strawberries in the way they influence the taste of the smoothie._ 
> 
> - _So, if there is no **interaction**, then the effect of adding more bananas on the taste of the smoothie will always be the same, no matter how many strawberries you put in. So the effect of bananas on the smoothie is the same whether you add a lot of strawberries or just a few: $\beta_b b_i + \beta_s s_i$_
> - _Or, on the other hand, if there is an **interaction**, then the effect of adding bananas (on the smootie flavor) will be different depending on how many strawberries there currently are in the smoothie: $\beta_b b_i + \beta_s s_i + \beta_{bs} (b_i\times s_i)$_
> 
> _In this case, the right answer is probably that the **linear form** with the **interaction** is correct. This is because the flavor probably depends on the relative amount of bananas and strawberries in the smoothie; so, the effect of adding a fixed amount of bananas to the smoothie probalby depends on the absolute amount of strawberries that are in the smoothie._
> 
> _Again, because understanding **interactions** in the context of **linear forms** is somewhat subtle and tricky and indeed not necessarily obviously intuitive, let's think about this a bit more. And we can simplify the concept a little bit by considering how this **interaction** would actually technically work in a **linear form** if we just had **binary indicator variables**._
>         
> - _To consider the smootie example in terms of binary variables, suppose that if both fruits are added to the smootie, they will be added in the same amount. So the smoothie will be made with either just bananas, just strawberries, or both (or neither and you won't make a smoothie)._ 
>     - _The question regarding an **interaction** then is, is the influence of the ingredients on the taste of the smoothie **additive** or **synergistic**? That is, does the way bananas affects the flavor of the smoothie change depending on the inclusion or exclusion of strawberries in the smoothie?_
>     - _**Additive** $\beta_b 1_{[b_i=1]}(b_i) + \beta_s 1_{[s_i=1]}(s_i)$ means there are three different flavors but they are explained by just two **parameters**: banana $\beta_b$, strawberry $\beta_s$, and banana-strawberry $\beta_b+\beta_s$_
>     - _**Synergistic** $\beta_b 1_{[b_i=1]}(b_i) + \beta_s 1_{[s_i=1]}(s_i) + \beta_{bs}(1_{[b_i=1]}(b_i) \times 1_{[s_i=1]}(s_i))$ means there are of course again three different flavors, but this time they are explained by three **parameters**: banana $\beta_b$, strawberry $\beta_s$, and banana-strawberry $\beta_b+\beta_s + \beta_{bs}$, which indicates that the flavor is "more than just sum of its parts", meaning there is a **synergistic interaction** and there is an interesting interplay between bananas and strawberries in the way they influence the taste of the smoothie_
>     
> _As the **additive** and **synergistic** versions of the **linear form** of the two **binary indicator variables** context shows, we don't need an interaction to make different predictions for different combinations of things. Instead, what these show is that the prediction will either be **additive** and "just the sum of it's parts" or **synergistic** (**interactive**) and "more than just sum of its parts"._

> _As introduced in the previous homework, **R-squared** is_
>
> _$$R^2 = 1 - \frac{\sum_{i=1}^n(y_i-\hat y)^2}{\sum_{i=1}^n(y_i-\bar y)^2}$$_
>    
> _which describes the **explanatory power** of a model; whereas, **p-values** allow us to characterize **evidence against** a **null hypothesis**, and **coefficients** in a **multiple linear regression** context allow us to interpret the relationship between the **outcome** and a **predictor variable** "with all other **predictor variables** 'held constant'". |p-value|Evidence|
|-|-|

|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|

> _In `formula='HP ~ Q("Sp. In this case we say the model is **overfit** (to the data its fit was based on). The purpose of using different **training** and **testing** datasets is to consider "in sample" versus "out of sample" **model performance** in order to try to confirm that the model is not **overfit** and that the **predictions** do indeed seem to **generalizable** beyond the dataset used for **model fitting**._

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import statsmodels.formula.api as smf

# Define model formula
model_formula = 'HP ~ Attack + Defense + Speed + Legendary + Q("Sp. Def") + Q("Sp. Atk")'

# Initialize lists to store R-squared values
in_sample_r2 = []
out_of_sample_r2 = []

# Perform 50 iterations with different train-test splits
for _ in range(50):
    # Split data (no fixed random seed in each loop iteration)
    pokeaman_train, pokeaman_test = train_test_split(pokeaman, train_size=0.5)
    
    # Fit the model on training data
    model_spec = smf.ols(formula=model_formula, data=pokeaman_train)
    model_fit = model_spec.fit()
    
    # Get in-sample R-squared
    in_sample_r2.append(model_fit.rsquared)
    
    # Calculate out-of-sample R-squared
    yhat_test = model_fit.predict(pokeaman_test)
    y_test = pokeaman_test.HP
    out_of_sample_r2.append(np.corrcoef(y_test, yhat_test)[0, 1] ** 2)

# Visualize results
plt.figure(figsize=(10, 5))
plt.plot(in_sample_r2, label="In-sample R-squared", marker='o')
plt.plot(out_of_sample_r2, label="Out-of-sample R-squared", marker='x')
plt.xlabel("Iteration")
plt.ylabel("R-squared")
plt.legend()
plt.title("In-sample vs. Out-of-sample R-squared over Different Train-Test Splits")
plt.show()
````

This illustration sheds light on the stability and generalisability of the model by illustrating how in-sample and out-of-sample R-squared values change over various train-test splits. Model fit on training data is indicated by in-sample R-squared, whereas performance on unknown data is indicated by out-of-sample R-squared.

Each iteration adds diversity in splits by not setting the random seed, enabling the observation of generalisation consistency. Overfitting, in which the model's predictions mostly depend on particular training data, may be indicated if out-of-sample R-squared values are continuously lower or extremely variable. This exercise emphasises how crucial it is to evaluate the model's robustness across splits in order to guarantee dependable performance on fresh data.
